Today we are working on a problem involving a 2D grid of electric potential values, which can be determined from its boundary conditions using Laplace's equation. Here you can see the discretized grid of electric potential values. $V^i_j$ corresponds to the potential at row $i$, column $j$ of the grid.

Each cell, having its own potential $V^i_j$, is located at $x_i = i\Delta x$, $y_j = j\Delta y$.

<img src="https://raw.githubusercontent.com/pattihamerski/PH-36X-Public/main/images/4x4-grid.png"
     alt="4x4 example grid"
     width="500"
/>

To apply Laplace's equation to the grid above, we can begin by writing Laplace's equation in 2 dimensions:

$$\nabla^2 V = \frac{\partial^2 V}{\partial x^2} + \frac{\partial^2 V}{\partial y^2} = 0$$

Now we see some partial differentials. Just like the last few times we've seen partial differentials, we apply our old friend Finite Difference Methods to discretize the expression:

$$\nabla^2 V = \frac{V^i_{j-1}-2V^i_j+V^i_{j+1}}{\Delta x^2} + \frac{V^{i-1}_j-2V^i_j+V^{i+1}_j}{\Delta y^2} = 0$$

In writing Laplace's equation this way, we took advantage of how shifting $i$ or $j$ corresponds to the $x$ or $y$ directions, as shown in the diagram below.

<img src="https://raw.githubusercontent.com/pattihamerski/PH-36X-Public/main/images/Vij-neighbors.png"
     alt="Vij and its neighboring values, which differ by +/- 1 in each of i, j."
     width="200"
/>

Finally, if we set the grid spacing so that $\Delta x = \Delta y$, then we can combine some terms and factor out some constants. We are left with:

$$V^i_j - \frac{1}{4}\left(V^{i-1}_j + V^{i+1}_j + V^i_{j-1} + V^i_{j+1}\right) = 0$$

The above equation can be used to solve for the potential of **interior cells** of the grid, but the exterior cells (highlighted green above) are on the boundary. They are set to pre-determined potential values using the known boundary conditions.

See below for the given boundary values of the 4-by-4 grid.

<img src="https://raw.githubusercontent.com/pattihamerski/PH-36X-Public/main/images/4x4-boundaries.png"
     alt="Known boundary values of the 4x4 grid. The center 4 values are unknown."
     width="300"
/>

From this point, we can write an entire system of equations, one for every cell in the grid:

$$
\begin{align}
V^0_0 &= 6 \\
V^0_1 &= 7 \\
V^0_2 &= 11 \\
V^0_3 &= 13 \\
V^1_0 &= 5 \\
V^1_1 - \frac{1}{4}\left(V^0_1 + V^2_1 + V^1_0 + V^1_2\right) &= 0 \\
V^1_2 - \frac{1}{4}\left(V^0_2 + V^2_2 + V^1_1 + V^1_3\right) &= 0 \\
V^1_3 &= 15 \\
V^2_0 &= 6 \\
V^2_1 - \frac{1}{4}\left(V^1_1 + V^3_1 + V^2_0 + V^2_2\right) &= 0 \\
V^2_2 - \frac{1}{4}\left(V^1_2 + V^3_2 + V^2_1 + V^2_3\right) &= 0 \\
V^2_3 &= 11 \\
V^3_0 &= 6 \\
V^3_1 &= 6 \\
V^3_2 &= 8 \\
V^3_3 &= 10
\end{align}
$$
