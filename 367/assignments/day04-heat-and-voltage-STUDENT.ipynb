In today's model, we apply a voltage across the wire. This introduces a new term to the heat equation we have been using. That terms comes from electrical power, defined like so and derived in part from Ohm's Law:

$$P = I\Delta V = \frac{\Delta V^2}{R}$$

Adding this term to our heat equation, and carefully including some constant factors to keep the units consistent, this is what we get:

$$\frac{\partial T}{\partial t} = \alpha \frac{\partial^2 T}{\partial x^2} + \frac{m}{c_P}\frac{\Delta V ^2}{R}$$

In this equation, $\alpha$ is thermal diffusivity, $m$ is mass, $c_P$ is specific heat capacity, $\Delta V$ is voltage drop, and $R$ is resistance. Some of these values can be looked up, but others are part of our model. One in particular, $R$, can be computed from properties of the material:

$$R = \rho \frac{L}{A}$$

where $\rho$ is electrical resistivity.
