Review the project description posted Canvas, and find peers who are interested in similar project topics as you. With your group, complete the following prompts.

#### &#9989; Project Goal

In one sentence, or one question, what physics problem do you want to solve or model?

**writing space here**

#### &#9989; Solution Path

What are the major steps you will need to take to achieve your goal? What are the anticipated milestones along the way to a solution?

**writing space here**

#### &#9989; Rough Outline of Coding

Summarize the main code chunks you think you will need to get to a solution. Create an outline of what code you expect you'll need to write. If you already have ideas for specific details of how your group will solve the problem, include those here. Existing ideas could include:
- specific coding structures
- past assignments to draw from
- any initial conditions
- existing data sets to use
- anything thing else already "known" about what your solution will look like!

**writing space here**

#### &#9989; Major Computing Tools/Concept Expected to be Used

Review your outline together. What are some of the individual computing skills or concepts that will be important to make use of? When will they be used in the project?

**writing space here**

#### &#9989; Plan of Work

As a reminder, we will have a few class days to work on the project later:
- Wed, 5/21: In-class project work time
- Wed, 5/28: In-class project work time
- Mon, 6/2: In-class project work time
- Wed, 6/4: Project presentations -- present solution as a group, each person expected to discuss the specific tool they focused on as part of the solution as a whole

With this in mind, please **discuss the following questions with your group and write your answers.**
- What are the expectations for each member in terms of participation?
- How will you communicate with one another about the group project? (even if someone has to miss class unexpectedly)
- What is the role of respect, cooperation, mutual trust, and openness in this group?
- If group members disagree on how to approach the group project, how will that conflict be handled?
- How will the group collaborate on the project and presentation, while still ensuring individuals have time to work on their computing tool review?
