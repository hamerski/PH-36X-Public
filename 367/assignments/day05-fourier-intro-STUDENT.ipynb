Last term in PH 366, we used Taylor Series to approximate values of a physical quantity close to a known value. More formally, if we had a known value $f(x_0)$, and the function $f$ (representing a physical quantity) was smooth near the point $x_0$, then we could approximate nearby values $f(x)$.

This approximation was useful for computing analytically complex quantities, like the electric potential near a symmetric distribution of point charges. We expressed Taylor Series as a sum of $N$ terms, which gets more and more precise as $N \rightarrow \infty$.

$$f(x) = \sum_{n=0}^\infty \frac{f^{(n)}(x_0)}{n!} (x-x_0)^n $$

Practically, Taylor Series could be applied to problems as shown the table below.

| | Taylor Series |
|-|-|
| **Formula** | $ f(x) = \sum_{n \ge 0} \frac{f^{(n)}(x_0)}{n!} (x-x_0)^n $ |
| **Starting info needed** | $x_0$ <br> $f^{(n)}(x_0)$ for $n = 0, 1, 2, 3,$ ... <br> values of $x$ centered around $x_0$ |
| **Values of $x$** | any values within radius of convergence, but better approximations for $x$ closer to $x_0$ |
| **Values of $f(x)$ computed at each step** | no steps; computation occurs all at once |
| **Code format** |  `approx = np.zeros(len(x_vals))`<br>` `<br>`for n in range(0, N):`<br>`    term = fn(x0, n) / math.factorial(n) * (x_vals - x0) ** n`<br>`    approx += term` |
| **Ideal usage** | Approximations physically close a point where derivatives of all orders are known at a single point |

Note in particular **the need for $x$ to be close to $x_0$.** As $x$ gets further away, higher values of $N$ are needed to make a precise approximation. Even further away, some values of $x$ cannot be approximated at all, because they lie outside the radius of convergence. ## Part 1: A series approximation with extended precision

Today, we begin to address the need for a precise approximation across a broad range of values. Unlike Taylor series, this new approximation does not require computing only the values close to some anchored $x_0$ value. However, *like* Taylor series, this approximation is a series sum, and it will become more and more precise as you increase the number of terms included.

&#9989; **Task 1.1:** Below, a generalized implementation of this new approximation is written in code. Review the code, and **write a formula** for this series sum.

```python
import numpy as np

# given terms
an = [25.37, 4.16, -2.24, 4.48, -2.07, 3.04, 1.56, -1.22, 1.33, -1.09]
bn = [0, -5.95, -2.11, 0.19, -2.09, -0.67, 1.78, 1.13, 0.16, -0.06]

# independent variable
x_vals = np.linspace(0, 40, 1000)

# series approximation
y_vals = np.zeros(len(x_vals))
y_vals += an[0] / 2

L = x_vals[-1] - x_vals[0]
N = len(an)

for n in range(1, N):
    a_term = an[n] * np.cos(2 * np.pi * n * x_vals / L)
    b_term = bn[n] * np.sin(2 * np.pi * n * x_vals / L)
    y_vals += a_term + b_term
```

## Part 3: Fourier series formula

The full Fourier series formula, which should look similar to your answer to Task 1.1, is:

$$f(x) = \frac{a_0}{2} + \sum_{n=1}^N a_n \cos\left(\frac{2\pi nx}{L}\right) + b_n \sin\left(\frac{2\pi nx}{L}\right)$$

This comes from splitting up the function $f(x)$ into a sum of sinusoidal terms, $a_n \cos\left(\frac{2\pi nx}{L}\right)$ and $b_n \sin\left(\frac{2\pi nx}{L}\right)$, for any integer $n$.

&#9989; **Task 3.1:** The $n=0$ term is unique -- there is no $b_0$ component, there is no $\sin$ or $\cos$, and the $a_0$ value itself is divided by 2. Discuss these unique traits with your peers, and come up with ideas for why this term is different from the others in this way.
