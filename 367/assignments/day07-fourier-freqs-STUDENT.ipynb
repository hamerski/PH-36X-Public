In the previous assignment, we ended by considering how the $a_n$ and $b_n$ coefficients of the Fourier series could be calculated directly from the musical note data.

$a_n$ and $b_n$ were defined like this:

$$a_n = \frac{2}{T}\int_0^T \cos\left(\frac{2\pi n t}{T}\right)f(t) \text{d}t$$

$$b_n = \frac{2}{T}\int_0^T \sin\left(\frac{2\pi n t}{T}\right)f(t) \text{d}t$$

Computing these values is useful because it allows us to see the relative contributions of different values of $n$ (i.e., different frequencies) on the overall Fourier series sum.

Consider the example shown in the the slides. If we had a perfect recording of a tone at 250 Hz, we would expect every value of $a_n$ and $b_n$ to be 0 *except* for the value of $n$ that corresponds to 250 Hz. 

Doing the conversion between $n$ and frequency can be accomplished by analyzing the input to our trigonometric functions:

$$\frac{2\pi n t}{T} = 2\pi \cdot \text{freq} \cdot t$$

Solving for frequency, we can show it is equal to $n/T$, where $T$ is the time period of the entire recording.

&#9989; **Task 0.1:** The function below is an example of a way to write the function from Task 2.1 from the Day 6 assignment, to calculate the $a_n$, $b_n$ coeefficients. **Add documention** to explain the purpose, inputs, and output of the function, **or just replace it** with your own function and documentation.

```python
# add proper documentation to the function below, or replace with your own

import numpy as np
from scipy import integrate

def anbn(f_vals, t_vals, n):
    T = t_vals[-1]
    trig_input = 2 * np.pi * n * t_vals / T
    an = 2 / T * integrate.trapezoid(np.cos(trig_input) * f_vals, x=t_vals)
    bn = 2 / T * integrate.trapezoid(np.sin(trig_input) * f_vals, x=t_vals)
    return an, bn
```
