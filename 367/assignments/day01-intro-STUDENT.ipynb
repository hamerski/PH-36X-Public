&#9989; **Task 2.2:** Together with your group, review the code below, which updates the temperature values across the entire wire over many time steps. Using what you see in the code, **write a formula for $\text{d}T_i/\text{d}t$**, the time-derivative of the temperature value for a single discretized piece of the wire.

**/your answer here/**

```python
# Code to review

t = 0
tf = 43200
dt = 1
dx = x[1] - x[0]

alpha = 1.27e-4

T = np.copy(T0)
N = len(T)

while t < tf:
    rate = np.zeros(N)
    for i in range(1, N-1):
        rate[i] = alpha * (T[i-1] - 2 * T[i] + T[i+1]) / dx**2
    dT = rate * dt
    T += dT
    t += dt
```
