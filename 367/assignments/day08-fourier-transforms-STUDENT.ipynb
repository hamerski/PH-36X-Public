## Part 1: Doing a "Fast" Fourier Transform

The idea behind Fourier transforms is that they transform the original function from the time domain ($t$) to the frequency domain ($\omega$). This is especially useful when analyzing data like the musical notes we have been working with, for with frequency is crucial.

The reason Fourier transforms can be so much faster is because they take advantage of the finite nature of digitized data. Each audio file stores a finite number of amplitude values. These values are all extracted from the audio file every time a single value of $a_n$ or $b_n$ is computed, which can make that process cumbersome for the computer's processor.

Fast Fourier transforms **only extract these values once**. Furthermore, they construct the entire freqeuncy distribution **in parallel computations,** rather than one term at a time. These two improvements work together to result in a highly efficient tool for computing Fourier frequencies.
