## Part 1: Solving systems of equations

Now let's put matrices to use! Consider this system of equations:

$$7x + 2y = 8$$
$$4x + 5y = 10$$

This can be rewritten in matrix form:

\begin{equation*}
\begin{pmatrix}
    7 & 2 \\
    4 & 5 
\end{pmatrix} \begin{pmatrix}
    x \\
    y
\end{pmatrix} = \begin{pmatrix}
    8 \\
    10
\end{pmatrix}
\end{equation*}

To solve a matrix equation $\textbf{A}\textbf{v} = \textbf{b}$ for the vector $\textbf{v}$, we can apply the inverse of $\textbf{A}$ to both sides:

$$\textbf{A}^{-1}\textbf{A}\textbf{v} = \textbf{A}^{-1}\textbf{b}$$

Since $\textbf{A}^{-1}\textbf{A}$ is the identity, this simplifies to: 

$$\textbf{v} = \textbf{A}^{-1}\textbf{b}$$
