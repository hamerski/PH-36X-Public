Last class, we computed the Fourier series approximation of a mystery function. Here is the code we used, based on given values for the `an` and `bn` terms:

```
x_vals = np.linspace(0, 40, 1000)

y_vals = np.zeros(len(x_vals))
y_vals += an[0] / 2

L = x_vals[-1] - x_vals[0]
N = len(an)

for n in range(1, N):
    a_term = an[n] * np.cos(2 * np.pi * n * x_vals / L)
    b_term = bn[n] * np.sin(2 * np.pi * n * x_vals / L)
    y_vals += a_term + b_term
```

Written as a mathematical formula, Fourier series can be represented like this:

$$f(x) = \frac{a_0}{2} + \sum_{n=1}^N a_n \cos\left(\frac{2\pi nx}{L}\right) + b_n \sin\left(\frac{2\pi nx}{L}\right)$$

This means some physical quantity $f(x)$, defined as a function of an independent variable $x$, can be approximated as a series sum of sinusoidal functions. The cool thing about Fourier series is that they can be applied to **any function**. At first glance, this is an amazing and powerful. It is, indeed, quite amazing and powerful.

The terms $a_n$ and $b_n$ each get multiplied with sines and cosines of $2\pi nx/L$, which are sunisoidal functions that have a frequency of $L/n$. In this way, the values of the $a_n$ and $b_n$ terms **indicate which frequencies contribute most strongly** to the overall function $f(x)$.

Today, we will use this principle to analyze a data set where frequency is vitally important to the underlying physics phenomenon: musical notes.

## Part 2: Constructing Fourier series

We now have $f(t)$, the amplitude as a function of time, for each instrument. We can use this to create Fourier series and see which frequencies are most salient for the two different musical notes. Remember, the $a_n$ and $b_n$ values can reflect the dominant frequencies in a function.

$a_n$ and $b_n$ are defined like this for any given function $f(t)$:

$$a_n = \frac{2}{T}\int_0^T \cos\left(\frac{2\pi n t}{T}\right)f(t) \text{d}t$$

$$b_n = \frac{2}{T}\int_0^T \sin\left(\frac{2\pi n t}{T}\right)f(t) \text{d}t$$

Note, now that we are working with time as our independent variable, the $x$ from before has been replaced with $t$ (time value), and the $L$ from before has been replaced with $T$ (time period).

The formulas above can be confirmed by plugging in the full series formula for $f(t)$ and simplifying the resulting integral.
