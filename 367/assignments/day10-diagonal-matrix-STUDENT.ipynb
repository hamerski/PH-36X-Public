In determining the force from the springs on each mass in terms of $k$, $m_0$, $m_1$, $\Delta x_0$, and $\Delta x_1$, we could do this a couple of different ways.

**First**, by directly computing the force from each spring. For example, $m_0$ can be pushed or pulled by two springs.
- The left spring, attached to the wall, will be compressed or stretched based on the value of $\Delta x_0$. This makes computing the force relatively simple: $F_{\text{left spring} \rightarrow m_0} = -k \Delta x_0$.
- The middle spring, also attached to $m_0$, is stretched/compressed depending on both $\Delta x_0$ and $\Delta x_1$. Positive values of $\Delta x_0$ will compress the spring, and positive values of $\Delta x_1$ will stretch it. The resulting force is then $F_{\text{middle spring} \rightarrow m_0} = -k (\Delta x_0 - \Delta x_1)$.
- In total, this yields: $F_{0\text{, springs}} = -2k\Delta x_0 + k\Delta x_1$.

**Second**, we could compute the force that results from *moving each mass* while keeping the others stationary.
- Moving $m_0$ itself can stretch and compress both the left spring and the middle spring. Both springs will push/pull the block in the direction of its $\Delta x_0=0$ position, so the force contribution is $F_{0\text{, moving } m_0} = -2k\Delta x_0$.
- Moving $m_1$ itself can stretch and compress the middle spring only (the right spring also moves, but this does not affect $m_0$). The middle spring will pull $m_0$ to the right when it is stretched in the same direction by $m_0$, so the force contribution is $F_{0\text{, moving } m_1} = k\Delta x_1$.
- In total, this once again yields: $F_{0\text{, springs}} = -2k\Delta x_0 + k\Delta x_1$.

Both approaches arrive at the same result. A similar process can be applied to $m_1$. The corresponding matrix equation is: 

\begin{equation*}
\begin{pmatrix}
    F_0 \\
    F_1 
\end{pmatrix} 
= \begin{pmatrix}
    -2k & k \\
    k & -2k
\end{pmatrix}
\begin{pmatrix}
    \Delta x_0 \\
    \Delta x_1
\end{pmatrix}
\end{equation*}

We will call the middle 2x2 matrix the "$\mathbb{K}$ matrix" in this assignment.
